# **Convergence of the Overlap Integral S(R=2a0)**
This plot shows the convergence of the Manto Carlo estimate overlap integral when R=2a0 and N, number of samples, increases. This plot uses both random sampling and importance sampling.

## Observations
- For the random sampling, the S(R) values fluctate intensely at smaller values of N and stabilize when N is at 10^5.
- For the importance sampling, the S(R) values have noticebaly smaller variance at lower N values (N=10^2-10^3). It begins to stabilize at N=10^6.

## Conclusion
- The average noise from the random sampling is greater than the average noise in the importance sampling, but the random sampling arrived at its stable value faster.
- The importance sampling is more efficient because the average noise is lower. With larger number of samples N, importance sampling will arrive at a stablize value noticably faster and clearer than random sampling.

# **Dependence of S(R) on Seperation Distance R=a0**
This plot shows the overlap intergral S(R) as 2 orbitals are seperated along the z-axis. 

## Observations
- At smaller values of R, S(R) is positive which shows a strong constructive overlap between the 2 orbitals.
- At R = ~4a0, S(R) crosses 0 and becomes negative. Between ~5a0 and ~15a0, the opposite-phase lobes of the orbitals overlap stronger than the same-phase ones.
- At R > 15a0, the overlap integral approaches 0 again because the orbitals get too far away from each other to interact.

## Conclusion
- At smaller values of R, the overlap integral corresponds to constructive interference between the orbitals. After R = ~4a0, the overlap integral is tracking the destructive interference of the orbitals.
